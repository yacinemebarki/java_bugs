```python
import pandas as pd
```

```python
df=pd.read_excel("22 Programming Languages Data.xlsx")
df
```

```python
print(df["code_preview"].iloc[50])
```

```python
print(df["language"].unique())
```

```python
language = [
    'Python','JavaScript','C++','C','Java','C#','Go','Ruby','PHP','Rust',
    'TypeScript','Kotlin','Swift','Scala','R','Shell','Perl','Objective-C',
    'Haskell','Lua','HTML','CSS'
]

for l in language:
    count=df[df["language"]==l].shape[0]
    print(f"{l}:{count}")
```

```python
df.info()
```

```python
df=df.dropna()
df.info()
```

```python
df=df[["language","code_preview"]]
df
```

```python
df.reset_index()
```

```python
print(df["code_preview"].iloc[20])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(df["code_preview"], df["language"], test_size=0.2, random_state=42)

model = Pipeline([
    ("tfidf", TfidfVectorizer(analyzer="char_wb", ngram_range=(2,5))), 
    ("clf", LogisticRegression(max_iter=300, multi_class="ovr"))
])

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
print(model.predict(['''public class EpochExample {
    public static void main(String[] args) {
        int dataSize = 10;   
        int batchSize = 2;   
        int epochs = 3;     

        
        for (int epoch = 1; epoch <= epochs; epoch++) {
            System.out.println("Epoch " + epoch + "/" + epochs);

            
            for (int i = 0; i < dataSize; i += batchSize) {
                int end = Math.min(i + batchSize, dataSize);
                System.out.println("  Processing batch: " + (i+1) + " to " + end);
            }
        }
    }
}''']))
```

```python
import joblib
joblib.dump(model, "model.pkl")
```
